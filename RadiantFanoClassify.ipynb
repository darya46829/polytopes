```python
import multiset
```

```python
def P3(n, k, m):
#partition of natural n into k natural numbers not more than m
    if n < 0 or m <= 0 or k < 0:
        yield None
    elif k == 0:
        if n == 0:
            yield multiset.Multiset()
        else:
            yield None
    
    else:
        for elem in range(m, 0, -1):
            for ms in P3(n - elem, k - 1, elem):
                if ms != None:
                    yield ms + {elem}

def P2(n, k):
#partition of natural n into k natural numbers
    for partition in P3(n, k, n - k + 1):
        yield partition

def P1(n):
#partition of natural n into natural numbers
    for k in range(1, n + 1):
        for partition in P2(n, k):
            yield partition
```
