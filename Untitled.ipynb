```python
def Get(filename, property_name):
    with open(filename) as file:
        string = '  <property name="' + property_name + '">\n'
        line = file.readline()
        while line and line != string:
            line = file.readline()

        if not line:
            raise Exception

        file.readline()
        vertices = []
        num = file.readline().split()
        while '</m>' not in num:
            num[-1] = num[-1][:-4]
            num = [int(coord) for coord in num[1:]]
            vertices.append(num)
            num = file.readline().split()
        return vertices
```

```python
def FaceToBasis(polyhedron):
    dim = polyhedron.dim()
    for facet in polyhedron.facets():
        f = facet.vertices()
        if len(f) > dim: continue
        A = matrix(f).transpose() ** -1
        yield facet, A

def FaceToBasisIsom(polyhedron):
    for facet, A in FaceToBasis(polyhedron):
        if abs(A.det()) == 1:
            yield facet, (A * matrix(polyhedron.vertices()).transpose()).transpose()
        else:
            raise Exception("Not Fano Polytope")

def Bilateral(rays):
    for ray in rays:
        negorth = true
        basis = true
        summa = 0
        for c in ray:
            summa += c
            if c < 0 or c > 1 or summa > 1: basis = False
            if c > 0: negorth = False
        if summa == 0 or not (negorth or basis): return False
    return True

def SpecialEmbedding(rays):
    return all(c >= 0 for c in sum(rays))
```

```python
col = vector([-1 for _ in range(30)])
def Dual(vert, faces):
    rays = matrix([vert[f].solve_right(col[:len(f)]) for f in faces])
    cones = [[] for v in vert]
    for f in range(len(faces)):
        for v in faces[f]:
            cones[v].append(f)
    return rays, cones
```

```python
fn = "fano-v4d/F.4D."
for i in range(0, FANO4D):
    if i < 10:
        filename = fn + "000" + str(i) + ".poly"
    elif i < 100:
        filename = fn + "00" + str(i) + ".poly"
    elif i < 1000:
        filename = fn + "0" + str(i) + ".poly"
    else:
        filename = fn + str(i) + ".poly"
    p = Polyhedron(vertices = Get(filename, "VERTICES")).polar()
    print("\npolyhedron", i, "\nvertices:\n", matrix(p.vertices()), "\nfacets:\n", [f.ambient_V_indices() for f in p.facets()])
    summa = 0
    for facet, polyhedron in FaceToBasisIsom(p):
        bil = Bilateral(polyhedron)
        se = SpecialEmbedding(polyhedron)
        if bil and se:
            summa += 1
        if not bil:
            continue
        print("facet ", facet.ambient_V_indices())
        print(polyhedron)
        print("\tBilateral: ", bil, "\tSpecial Embedding: ", se, "\tboth: ", bil and se)
    print("       ", summa)
```
